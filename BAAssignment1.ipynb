```python
Q1 = data7['requests'].quantile(0.25)
Q2 = data7['requests'].quantile(0.75)
IQR = Q2 - Q1
requests_lower_bound = Q1 - 1.5*IQR
requests_upper_bound = Q2 + 1.5*IQR
outliers = data7[(data7['requests'] < requests_lower_bound) | (data7['requests'] > requests_upper_bound)]
print(outliers)
```

```python
Q1 = data7['ae_responses'].quantile(0.25)
Q2 = data7['ae_responses'].quantile(0.75)
IQR = Q2 - Q1
ae_responses_lower_bound = Q1 - 1.5*IQR
ae_responses_upper_bound = Q2 + 1.5*IQR
outliers = data7[(data7['ae_responses'] < ae_responses_lower_bound) | (data7['ae_responses'] > ae_responses_upper_bound)]
print(outliers)
```

```python
Q1 = data7['ae_impressions'].quantile(0.25)
Q2 = data7['ae_impressions'].quantile(0.75)
IQR = Q2 - Q1
ae_impressions_lower_bound = Q1 - 1.5*IQR
ae_impressions_upper_bound = Q2 + 1.5*IQR
outliers = data7[(data7['ae_impressions'] < ae_impressions_lower_bound) | (data7['ae_impressions'] > ae_impressions_upper_bound)]
print(outliers)
```

```python
Q1 = data7['ae_clicks'].quantile(0.25)
Q2 = data7['ae_clicks'].quantile(0.75)
IQR = Q2 - Q1
ae_clicks_lower_bound = Q1 - 1.5*IQR
ae_clicks_upper_bound = Q2 + 1.5*IQR
outliers = data7[(data7['ae_clicks'] < ae_clicks_lower_bound) | (data7['ae_clicks'] > ae_clicks_upper_bound )]
print(outliers)
```

```python
Q1 = data7['ae_revenue'].quantile(0.25)
Q2 = data7['ae_revenue'].quantile(0.75)
IQR = Q2 - Q1
ae_revenue_lower_bound = Q1 - 1.5*IQR
ae_revenue_upper_bound = Q2 + 1.5*IQR
outliers = data7[(data7['ae_revenue'] < ae_revenue_lower_bound) | (data7['ae_revenue'] > ae_revenue_upper_bound)]
print(outliers)
```

```python
def ok(requests,ae_responses,ae_impressions, ae_clicks, ae_revenue,requests_lower_bound,requests_upper_bound,ae_responses_lower_bound, ae_responses_upper_bound,ae_impressions_lower_bound,ae_impressions_upper_bound,ae_clicks_lower_bound,ae_clicks_upper_bound, ae_revenue_lower_bound, ae_revenue_upper_bound) :
  i = 0
  if requests < requests_lower_bound or requests > requests_upper_bound :
    i = i + 1
  if ae_responses < ae_responses_lower_bound or ae_responses > ae_responses_upper_bound :
    i = i + 1
  if ae_impressions < ae_impressions_lower_bound or ae_impressions > ae_impressions_upper_bound :
    i = i + 1
  if ae_clicks < ae_clicks_lower_bound or ae_clicks > ae_clicks_upper_bound :
    i = i + 1
  if ae_revenue < ae_revenue_lower_bound or ae_revenue > ae_revenue_upper_bound :
    i = i + 1
  if i == 4 :
    return 'outlier'
  else :
    return 'nan'


   
  
```

```python
data7['status'] =  data7.apply(lambda x : ok(x.requests,x.ae_responses,x.ae_impressions, x.ae_clicks, x.ae_revenue,requests_lower_bound,requests_upper_bound,ae_responses_lower_bound , ae_responses_upper_bound,ae_impressions_lower_bound,ae_impressions_upper_bound,ae_clicks_lower_bound,ae_clicks_upper_bound, ae_revenue_lower_bound, ae_revenue_upper_bound),axis=1)
```
